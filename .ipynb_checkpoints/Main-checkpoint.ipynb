### Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline

# Dataframe settings
pd.set_option('display.max_columns', 50, 'display.max_rows', 500, 'display.float_format', '{:.2f}'.format)
```

### Import data

```python
# Import dataset
df_brancher = pd.read_csv("Data/arbejdsmarkedsanalyse_brancher.csv", sep=';', decimal=",", encoding='iso8859_10')
df_personer = pd.read_csv("Data/arbejdsmarkedsanalyse_koen_alder.csv", sep=';', decimal=",", encoding='iso8859_10')
```

### Cleanse data

```python
# ----- Cleanse "Brancher" -----
# Remove "Total" as it is not group by itself
df_brancher = df_brancher[df_brancher["Group"] != "Total"]
```

```python
# ----- Cleanse "Køn og Alder" -----
# Split "Group" in two columns
df_personer[['Køn', 'Aldersgruppe']] = df_personer.Group.str.split(pat=',', expand=True)
df_personer = df_personer.drop(columns=['Group'])
```

### Descriptive statistics - "Brancher"

```python
# Show data preview
df_brancher.head(3)
```

```python
# Get general info about the dataframe
df_brancher.info()

# Get descriptive statistics
df_brancher.describe(include='all')
```

### Visualize data - "Brancher"

```python
# ----- Number of respondents by "Group" -----

# Get number of respondents by "group"
count_group = df_brancher.groupby(["Group"])["Antpers"].max()

# Make into dataframe for visualizations purposes, and sort by number of respondents
count_group = pd.DataFrame(count_group).sort_values("Antpers", ascending=False)

# Plot number of respondents by "Group"
sns.set_theme(style="darkgrid")
plt.subplots(figsize=(5, 15))

sns.barplot(x=count_group.Antpers, y=count_group.index, data=count_group, palette="GnBu_r")
plt.show()
```

```python
# ----- Number of respondents by "Topic Label" -----

# Get number of respondents by "Topic Label"
count_label = df_brancher.groupby(["Topic Label"])["Antpers"].sum()

# Make into dataframe for visualizations purposes, and sort by number of respondents
count_label = pd.DataFrame(count_label).sort_values("Antpers", ascending=False)

# Plot number of respondents by "Topic Label"
sns.set_theme(style="darkgrid")
plt.subplots(figsize=(5, 15))

sns.barplot(x=count_label.Antpers, y=count_label.index, data=count_label, palette="GnBu_r")
plt.show()
```

### Finding 1
- __Indikation 1: Brancher med meget manuelt fysisk arbejde har typisk et værre arbejdsmiljø eller har medarbejdere med dårligere helbred__  

- __Indicakion 2: Brancher bestående af større virksomheder har typisk et bedre arbejdsmiljø eller har medarbejdere med bedre helbred__

```python
# ----- Plot Groups by their score -----

# Calculate mean of the indexed scores
groupScore = pd.DataFrame(df_brancher.groupby(["Group"])["Score (Indekseret score)"].mean().sort_values(ascending=False))

# Create plot
sns.set_theme(style="darkgrid")
fig, ax = plt.subplots(figsize=(5, 15))

sns.barplot(ax=ax, x=groupScore["Score (Indekseret score)"], y=groupScore.index, data=groupScore, palette="GnBu_r")
ax.set_xlabel("Gns. indekseret score", size=14)
ax.set_ylabel("Branche", size=14)
ax.set_title("Gennemsnitlig indekseret score per branche", size=18, pad=18)
plt.show()
```

### Finding 2
- __Psykisk arbejdsbelastning er et mere hyppigt problem end fysisk arbejdsbelastning.__

```python
# Create dictionary that maps question as psychological or physical (or undefineable)
isPsychological = dict((element,0) for element in df_brancher["Topic Label"].unique())

# Manually label psychological types
isPsychological["Angstsymptomer"] = 1
isPsychological["Arbejdsevne"] = "Return" # <- Needs to further manually specified as it holds both types
isPsychological["Arbejdspladsens prioritering af arbejdsmiljø"] = "Undefineable"
isPsychological["Arbejdsrelateret sygdom"] = "Undefineable"
isPsychological["Arbejdstid"] = "Undefineable"
isPsychological["Depressive symptomer"] = 1
isPsychological["Egen indsats for arbejdsmiljø"] = 1
isPsychological["Engagement og mening i arbejdet"] = 1
isPsychological["Følelsesmæssige krav"] = 1
isPsychological["Helbred"] = "Undefineable"
isPsychological["Indflydelse"] = "Undefineable"
isPsychological["Kollegial støtte og samarbejde"] = 1
isPsychological["Jobusikkerhed"] = 1
isPsychological["Kollegial anerkendelse"] = 1
isPsychological["Konflikter mellem arbejde og privatliv"] = 1
isPsychological["Kvantitative krav og grænseløshed"] = 1
isPsychological["Ledelseskvalitet"] = 1
isPsychological["Mentalt helbred"] = 1
isPsychological["Negative oplevelser"] = 1
isPsychological["Nyansatte"] = "Undefineable"
isPsychological["Retfærdighed"] = 1
isPsychological["Rolleklarhed"] = 1
isPsychological["Rummelighed"] = 1
isPsychological["Sundhedsfremmetilbud"] = "Undefineable"
isPsychological["Træthed og søvnproblemer"] = 1
isPsychological["Uoverskuelighed og stress"] = 1
isPsychological["Vold og trusler"] = "Return"
isPsychological
```

```python
# ----- Insert values into main dataframe -----

# Create default values
df_brancher["isPsychological"] = 0

# Insert proper values
for key in isPsychological:
    df_brancher.loc[df_brancher["Topic Label"] == key, "isPsychological"] = isPsychological[key]
    
# Manually insert values for special cases
df_brancher.loc[df_brancher["Question Label"] == "Fysisk arbejdsevne", "isPsychological"] = 0
df_brancher.loc[df_brancher["Question Label"] == "Arbejdsevne", "isPsychological"] = "Undefineable"
df_brancher.loc[df_brancher["Question Label"] == "Psykisk arbejdsevne", "isPsychological"] = 1
df_brancher.loc[df_brancher["Question Label"] == "Fysisk vold", "isPsychological"] = 0
df_brancher.loc[df_brancher["Question Label"] == "Trusler", "isPsychological"] = 1
```

```python
# ----- Do calculations on physical/psychological related questions -----

# Calculate mean score on Fysical/Psychological respectively
scoreByType = df_brancher.groupby(["Group", "isPsychological"])["Score (Indekseret score)"].mean().unstack()

# Rename columns
scoreByType = scoreByType.rename(columns={0: "Fysisk", 1: "Psykologisk"})

# Add number of respondents by "Branche"
scoreByType['Count'] = df_brancher.groupby(["Group"])["Antpers"].max()

# Add overall score
scoreByType['Overall score'] = df_brancher.groupby(["Group"])["Score (Indekseret score)"].mean()

# Remove 'Undefineable' cases 
scoreByType.drop(columns=['Undefineable'])
scoreByType
```

```python
# ----- Plot results -----
sns.set_theme(style="darkgrid")

g = sns.relplot(data=scoreByType,
                x='Fysisk', 
                y='Psykologisk',
                hue='Overall score',
                size="Count",
                palette="GnBu",
                sizes=(10, 400))

# Insert group names
for x, y, i in zip(scoreByType['Fysisk'], scoreByType['Psykologisk'], scoreByType.index):
     plt.text(x = x+.2,
              y = y-.03,
              s = i, 
              color = 'purple',
              size=4)

g.fig.suptitle("Fysisk versus psykisk belastning i arbejdsmiljøet (per branche)", size=16, y=1.02)
g.savefig("Output/Finding_2.png", dpi = 300)
plt.show() # <- Please visit the saved image at ../Output/
```

### Finding 3
- __Offentligt ansatte har færre tilbud om behandlingsordninger.__

```python
# Do calculations on "Behandlingsordning (bh)"
bh = df_brancher[df_brancher["Question Label"] == "Tilbud om behandlingsordning"][["Group", "Score", "Antpers"]]
bh = bh.set_index("Group")
bh['Respondenter uden sundhedstilbud'] = bh.Antpers * (1-(bh.Score/100))
bh = bh.sort_values(by="Respondenter uden sundhedstilbud",ascending=False)
bh
```

```python
# Create plot
sns.set_theme(style="darkgrid")
fig, ax = plt.subplots(figsize=(5, 15))

sns.set_color_codes("muted")
sns.barplot(x=bh["Antpers"], y=bh.index, data=bh, 
            color="b", label="Total antal respondenter")

sns.set_color_codes("pastel")
sns.barplot(x=bh["Respondenter uden sundhedstilbud"], 
            y=bh.index, data=bh, color="b", label="Antal respondenter uden sundhedstilbud")# palette="GnBu_r")

ax.legend(ncol=1, loc="lower right", frameon=True)
ax.set_xlabel("Antal respondenter", size=14)
ax.set_ylabel("Branche", size=14)
ax.set_title("Antal respondenter uden sundhedstilbud", size=18, pad=18)
plt.show()
```

### Finding 4
- __Helbredet og arbejdsmiljøet er det samme på tværs af medarbejdernes køn og alder, psykisk såvel som fysisk.__

```python
# ----- Add type of questions (psychological/physical) -----

# Insert default values into dataframe
df_personer["isPsychological"] = 0

# Insert real values into dataframe
for key in isPsychological:
    df_personer.loc[df_personer["Topic Label"] == key, "isPsychological"] = isPsychological[key]
    
# Manually insert values for special cases
df_personer.loc[df_personer["Question Label"] == "Fysisk arbejdsevne", "isPsychological"] = 0
df_personer.loc[df_personer["Question Label"] == "Arbejdsevne", "isPsychological"] = "Undefineable"
df_personer.loc[df_personer["Question Label"] == "Psykisk arbejdsevne", "isPsychological"] = 1
df_personer.loc[df_personer["Question Label"] == "Fysisk vold", "isPsychological"] = 0
df_personer.loc[df_personer["Question Label"] == "Trusler", "isPsychological"] = 1
```

```python
# ----- Calculate mean score on Physical/Psychological respectively -----

# Create dataframe
scorePhyPsy = df_personer.groupby(["Køn", "Aldersgruppe", "isPsychological"])["Score (Indekseret score)"].mean().unstack()

# Rename columns
scorePhyPsy = scorePhyPsy.rename(columns={0: "Fysical", 1: "Psychological"})

# Drop undesired column
scorePhyPsy = scorePhyPsy.drop(columns=['Undefineable'])

# Add number of respondents
scorePhyPsy['Count'] = df_personer.groupby(["Køn", "Aldersgruppe"])["Antpers"].max()

# Add indexed score
scorePhyPsy['Score'] = df_personer.groupby(["Køn", "Aldersgruppe"])["Score (Indekseret score)"].mean()

# Calculate psychologicals proportion of the normal indexed score
scorePhyPsy['PsycProp'] = (scorePhyPsy.Psychological / (scorePhyPsy.Fysical + scorePhyPsy.Psychological)) * scorePhyPsy.Score

scorePhyPsy = scorePhyPsy.reset_index()
scorePhyPsy
```

```python
# ----- Plot distribution -----
sns.set_theme(style="darkgrid")

sns.set_color_codes("muted")
sns.barplot(data=scorePhyPsy, x='Aldersgruppe', y='Count', hue='Køn')

plt.show()
```

```python
# ----- Plot psychological proportion by sex and age -----

sns.set_theme(style="darkgrid")

sns.set_color_codes("muted")
sns.barplot(data=scorePhyPsy, x='Aldersgruppe', y='Score', hue='Køn')#, color="b")

sns.set_color_codes("pastel")
g = sns.barplot(data=scorePhyPsy, x='Aldersgruppe', y='PsycProp', hue='Køn', color="g")

# Define legends
handles, labels = g.get_legend_handles_labels()
labels = ['Kvinder (total score)', 'Mænd (total score)', 'Kvinder (psykologisk andel)', 'Mænd (psykologisk andel)']
g.legend(ncol=1, loc="lower right", frameon=True, handles=handles, labels=labels)

g.set_xlabel("Aldersgruppe", size=14)
g.set_ylabel("Score (indekseret)", size=14)
g.set_title("Psykologisk arbejdsbelastning som andel af den totale arbejdsbelastning", size=16, pad=18)
plt.show()
```
